```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# read the data
df1 = pd.read_csv("wine.csv")
df1.head()
```

```python
df1.describe()
```

```python
df1.info()
```

```python
df1.isnull().sum()
```

```python
#Remove duplicate data
df1 = df1.drop_duplicates()
df1
```

```python
#Replace Null values with mean value
df1["fixed acidity"].fillna(df1["fixed acidity"].mean(),inplace=True)

df1["volatile acidity"].fillna(df1["volatile acidity"].mean(),inplace=True)

df1["citric acid"].fillna(df1["citric acid"].mean() , inplace = True)

df1["residual sugar"].fillna(df1["residual sugar"].mean() , inplace = True)

df1["chlorides"].fillna(df1["chlorides"].mean() , inplace = True)

df1["pH"].fillna(df1["pH"].mean() , inplace = True)

df1["sulphates"].fillna(df1["sulphates"].mean() , inplace = True)
```

```python
df1.isnull().sum()
```

```python
#Visualize the data
plt.figure(figsize=(10,7))
plt.scatter(x="alcohol",y="fixed acidity",data =df1,marker= 'o',c="m")
plt.xlabel("alcohol",fontsize=15)
plt.ylabel("fixed_acidity",fontsize=15)
plt.show()
```

```python
#Linear plot
sns.lmplot(x="alcohol",y="fixed acidity",data=df1)
plt.plot()
```

```python
#Visualizing all attributes
sns.set()
sns.pairplot(df1)
plt.show()
```

```python
#Box Plot to identify outliers in the Data
sns.set()
plt.figure(figsize=(20,10))
sns.boxplot(data=df1,palette="Set3")
plt.show()
```

```python
#Calculate quantile values
Q1 = df1.quantile(0.25)
Q3 = df1.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
#Remove outliers from the data
print((df1 < (Q1 - 1.5 * IQR)) | (df1 > (Q3 + 1.5 * IQR)))
df2 = df1[(df1["total sulfur dioxide"] > (Q1[6] - 1.5 * IQR[6])) & (df1["total sulfur dioxide"] < (Q3[6] + 1.5 * IQR[6]))]
df3 = df2[(df2["free sulfur dioxide"] > (Q1[5] - 1.5 * IQR[5])) & (df2["free sulfur dioxide"] < (Q3[5] + 1.5 * IQR[5]))]
df4 = df3[(df3["residual sugar"] > (Q1[3] - 1.5 * IQR[3])) & (df3["residual sugar"] < (Q3[3] + 1.5 * IQR[3]))]
```

```python
#Clean data
sns.set()
plt.figure(figsize=(20,10))
sns.boxplot(data=df4,palette="Set3")
plt.show()
```

```python
#Modifying 'type' attribute
dummies = pd.get_dummies(df4["type"],drop_first=True)
df4 = pd.concat([df4,dummies],axis=1)
df4.drop("type",axis=1,inplace=True)
df4
```

```python
df4.quality.value_counts()
```

```python
#Renaming 'quality' attribute
quality_desc = { 3 : "Low",4 : "Low",5: "Medium",6 : "Medium",7: "Medium",8 : "High",9 : "High"}
df4["quality"] =  df4["quality"].map(quality_desc)
```

```python
df4.quality.value_counts()
```

```python
mapping_quality = {"Low" : 0,"Medium": 1,"High" : 2}
df4["quality"] =  df4["quality"].map(mapping_quality)
```

```python
df4.quality.value_counts()
```

```python
#feature selection for our model
from sklearn.ensemble import ExtraTreesClassifier
y = df4["quality"]
X = df4.drop("quality",axis=1)
model = ExtraTreesClassifier()
model.fit(X,y)
print(model.feature_importances_)
```

```python
plt.figure(figsize=(30,10))
plt.bar(X.columns, model.feature_importances_) 
plt.xlabel('Feature Labels') 
plt.ylabel('Feature Importances') 
plt.title('Comparison of different Feature Importances') 
plt.show() 
```

```python
#select the top 9 features
important_features = pd.Series(model.feature_importances_,index =X.columns)
print(important_features)
important_features = important_features.nlargest(9)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
#preparing the dataset
df5 = df4.loc[ : , ['volatile acidity','free sulfur dioxide','alcohol','residual sugar','fixed acidity','sulphates','chlorides','pH','density','quality']]
```

```python
df5.info()
```

```python
#dividing the data into features and class label
X1 = df5.drop("quality",axis=1)
Y1 = df5["quality"]
```

```python
#Splitting train and test data
x_train,x_test,y_train,y_test = train_test_split(X1,Y1,test_size=0.2,random_state=0)
```

```python
#Logistic Regression
clf1 = LogisticRegression(solver="liblinear",multi_class="auto")
clf1.fit(x_train,y_train)
y_pred = clf1.predict(x_test)
acc = accuracy_score(y_test,y_pred)
print(acc)
```

```python
#Random forest classifier
clf2 = RandomForestClassifier(n_estimators=10,max_depth=5)
clf2.fit(x_train,y_train)
y_pred2 = clf2.predict(x_test)
acc = accuracy_score(y_test,y_pred2)
print(acc)
```

```python
#Naive Bayes
clf3 = GaussianNB()
clf3.fit(x_train,y_train)
y_pred3 = clf3.predict(x_test)
acc = accuracy_score(y_test,y_pred3)
print(acc)
```

```python
#Decision Tree Classifier
clf3 = DecisionTreeClassifier(criterion="entropy",max_depth=5)
clf3.fit(x_train,y_train)
y_pred3 = clf3.predict(x_test)
acc = accuracy_score(y_test,y_pred3)
print(acc)
```

```python
#SVC
clf3 = SVC(C=1,kernel='rbf',gamma="scale")
clf3.fit(x_train,y_train)
y_pred3 = clf3.predict(x_test)
acc = accuracy_score(y_test,y_pred3)
print(acc)
```

# Out of all the models run on the data, Logistic Regression ,Random Forest Classifier and SVC gave 93% accuracy.

## Hence, Wine quality can be predicted with 93% accuracy
